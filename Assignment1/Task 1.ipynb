```python
import pandas as pd
import numpy as np
# from pathlib import Path # delete?
import missingno as msno
```

```python
def cleaner():
    filePath = "HealthSurvey/NHANES_age_prediction.csv"
    survey = pd.read_csv(filePath)  # Load csv
    
    print('3 Missing Info Matrices correspond to:')
    print('Survey, Breast Cancer (pre drop), Breast Cancer (post drop)')
    msno.matrix(survey) # Visualize any missing values
    survey = survey.drop(["RIDAGEYR", "LBXGLT"], axis=1) # Take out non-significant cols
    # survey.to_csv("assignment_data/HealthSurvey/NHANES_age_prediction_clean.csv")


    filePath = "BreastCancer/breast-cancer-wisconsin.data"
    breastCancer = pd.read_csv(filePath)
    breastCancer.columns = ["Sample_code_number", "Clump_thickness", "Uniformity_of_cell_size", "Uniformity_of_cell_shape", "Marginal_adhesion", "Single_epithelial_cell_size", "Bare_nuclei", "Bland_chromatin", "Normal_nucleoli", "Mitoses", "Class"]
    breastCancer = breastCancer.replace('?', np.nan)
    msno.matrix(breastCancer) # pre-drop
    breastCancer.dropna(inplace=True)
    msno.matrix(breastCancer) # post-drop
    # breastCancer.to_csv("assignment_data/BreastCancer/breast-cancer-wisconsin_clean.csv")
    breastCancer['Bare_nuclei']=breastCancer['Bare_nuclei'].astype(int) # Bare_nuclei converted from str to int

    return survey, breastCancer


survey, breastCancer = cleaner()
```

Notice the second and third matrices differ due to the dropped rows that have empty values. The first matrix does not have any empty values.

3. Compute basic statistics on the data to understand it better. For example, what are the mean of each feature for the positive group and the mean of each feature for the negative group? If you rank the squared difference of the group means, are the top features known to be associated with the target variable?

```python
# Class:
# 2=benign, set as negative group
# 4=malignant, set as positive group
print('features:', breastCancer.columns)
```

```python
def calc_feature_means():
    negativeGroup = breastCancer[breastCancer['Class']==2]
    positiveGroup = breastCancer[breastCancer['Class']==4]

    featureMeansNegativeGroup = []
    featureMeansPositiveGroup = []
    
    for col in negativeGroup.columns:
        featureMeansNegativeGroup.append(negativeGroup.loc[:, col].mean())
        featureMeansPositiveGroup.append(positiveGroup.loc[:, col].mean())
    squared_differences = [(neg-pos)**2 for (neg,pos) in zip(featureMeansNegativeGroup, featureMeansPositiveGroup)]
    
    rows = ['negative_group','positive_group','squared_differences']
    cols = breastCancer.columns
    data = [featureMeansNegativeGroup, featureMeansPositiveGroup, squared_differences]
    df_group_means = pd.DataFrame(data, index=rows, columns=cols)
    
    return df_group_means
    
statistical_analysis = calc_feature_means()
statistical_analysis
```

```python
# Sort features by increasing values of squared_differences
statistical_analysis.sort_values('squared_differences', axis=1, inplace=True)
statistical_analysis
```

NOTE: The row 'squared_differences' is calculated by finding (neg_group_mean - pos_group_mean)^2 for every feature/column
